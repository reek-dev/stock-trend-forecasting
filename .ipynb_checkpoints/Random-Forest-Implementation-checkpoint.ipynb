```python
import yfinance as yf

msft = yf.Ticker("MSFT")
msft = msft.history(period="max")
```

```python
msft.head()
```

```python
msft.plot.line(y="Close", use_index=True)
```

```python
data = msft[["Close"]]
data = data.rename(columns = {'Close' : 'Actual Close'})
data["Target"] = msft.rolling(2).apply(lambda x : x.iloc[1] > x.iloc[0])["Close"]
```

```python
data.head(20)
```

```python
msft_prev = msft.copy()
msft_prev = msft_prev.shift(1)
```

```python
msft_prev.head(10)
```

```python
predictors = ["Close", "High", "Low", "Open", "Volume"]
data = data.join(msft_prev[predictors]).iloc[1:]

data.head(10)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=200, random_state=1)

# first we are taking the entire dataframe except the last 100 rows as the train
train = data.iloc[:-100]

# then we are taking only the last 100 rows as the test
test = data.iloc[-100:]

# Build a forest of trees from the training set (X, y)
model.fit(X=train[predictors], y=train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
preds
```

```python
import pandas as pd
preds = pd.Series(preds, index=test.index)
preds
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat({"Target" : test["Target"], "Predictions" : preds}, axis=1)
combined
```

```python
combined.plot()
```

```python
start = 1000
step = 750
i = 1000

train = data.iloc[0:i].copy()
test = data.iloc[i: (i+step)].copy()

model.fit(X=train[predictors], y=train["Target"])


# now this will return an array of array with 2 elements in it
# the first element indicates the probability that the stock will go down
# the second element indicates the probability that the stock will go up
preds = model.predict_proba(test[predictors])
preds
```
